## Comparison of Employment Ratio for Women by Race

```python
# Specify the columns for all women by race
columns = [
    'year',
    'white_women',
    'black_women',
    'hispanic_women'
]

# Create a DataFrame with the selected columns 2012 to 2022
women_by_races = df[columns]
women_by_race = women_by_races.head(10)
women_by_race

```

```python
# Group by 'year' and calculate the mean employment ratio for each racial group
avg_employment_by_race = women_by_race.groupby('year').mean().reset_index()
avg_employment_by_race
```

```python
#Visualization

# Set the figure size
plt.figure(figsize=(4, 3))

# Create line plots for each racial group
sns.lineplot(x='year', y='white_women', data=avg_employment_by_race, label='White Women')
sns.lineplot(x='year', y='black_women', data=avg_employment_by_race, label='Black Women')
sns.lineplot(x='year', y='hispanic_women', data=avg_employment_by_race, label='Hispanic Women')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Average Employment to Population Ratio for Women',fontsize=6)
plt.title('Comparison of Employment Ratio for Women by Race(2012-2022)',fontsize=6)
plt.legend(fontsize=5)  

plt.show()
```
